### Question - Given a rectangle and a point. Find if the point lies inside the rectangle or outside it.

    Input: ([0, 2, 2, 1], [3, 4])
        In the above input, first list represents -> [Top left coordinates, width, height] and second list represent the given point.

    Output: Outside

```python
def locate_point(rectangle,point):
    x=point[0]
    y=point[1]
    
    x1=rectangle[0]
    y1=rectangle[1]
    
    x2=x1+rectangle[2]
    y2=y1-rectangle[3]
    
    if (x>=x1 and  x<=x2 and y<=y1 and y>=y2):
        print('inside')
    else:
        print('outside')
        
input_=[[0,2,2,1],[1,2]]
locate_point(input_[0],input_[1])
    
```

### Problem 2
Given a read-only array arr of size N. The property of this array is that,

    1. Each element of this array is going to fall in the range of 1 to N (where N is the size of array).
    2. All element comes exaclty once, except for 1 element. Only one element comes twice.

#### Output the repeated and the missing value?

    Input: arr = [3, 1, 2, 5, 3]  
    Output: 3 4

```python
# O(n^2)
arr=[3, 1, 2, 5, 3]

# arr = [3, 1, 2, 5, 3]
# repeated number
for i in range(len(arr)):
    for element in arr[i+1:]:
        if arr[i]==element:
            print(arr[i],end=' ')
# missing number
max_val=len(arr)

for i in range(1,max_val+1):
        if i not in arr:
            print(i)
```

```python
height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
# height = [3, 2, 1, 0, 4]

water_trapped = 0

max_left_arr = []
max_ = height[0]
for i in height:
  if max_ < i:
    max_ = i
  max_left_arr.append(max_)

max_right_arr = [ ]
max_ = height[-1]
for i in height[::-1]:
    if max_ < i:
        max_ = i
    max_right_arr.insert(0,max_)

for i in range(1, len(height) - 1):
  
  curr_bar_height = height[i]
  max_left = max_left_arr[i]
  max_right = max_right_arr[i]

  if (curr_bar_height < max_left and curr_bar_height < max_right):
    water_trapped += min(max_left, max_right) - curr_bar_height

print(water_trapped)
```
